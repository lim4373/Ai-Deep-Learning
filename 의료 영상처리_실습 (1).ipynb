## Correlation and Convolution

### 2d correlation
\begin{equation}
h [m, n] = \sum_{k,l} f [k, l] I [m+k, n+l] \qquad (1)
\end{equation}
The equation $(1)$ can be easily implemented by

<center> <code>
        h = scipy.signal.correlate2d(I, f)
</code> </center>

### 2d convolution
\begin{equation*}
h [m, n] = \sum_{k,l} f [k, l] I [m-k, n-l] \qquad (2)
\end{equation*}
It can be easily implemented by
<center> <code>
h = scipy.signal.convolve2d(I, f)
</code> </center>

```python
f = np.array([[0, 0, 0, 0, 0],
              [0, 1/9, 1/9, 1/9, 0],
              [0, 1/9, 1/9, 1/9, 0],
              [0, 1/9, 1/9, 1/9, 0],
              [0, 0, 0, 0, 0]])

f.shape
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,5))
b1= signal.correlate2d(brain3,f)
axs[0].set_title('barin_tumor, correlate2d')
axs[0].imshow(b1,cmap='bone')

axs[1].set_title('brain_tumor')
axs[1].imshow(brain3,cmap='bone')
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,5))
b1= signal.correlate2d(brain3,f)
axs[0].set_title('barin_tumor, correlate2d')
axs[0].imshow(b1,cmap='bone')

b2= signal.convolve2d(brain3, f)
axs[1].set_title('brain_tumor,convolve2d')
axs[1].imshow(b2,cmap='bone')
```

```python
import skimage
fig, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_title("brain_tumor")
axs[0].imshow(brain, cmap='gray')


D = skimage.img_as_float(b1)
f1= D[70:170, 20:90]
axs[1].set_title("brain_tumor_segmentation_correlate2d")
axs[1].imshow(f1, cmap='bone')
```
