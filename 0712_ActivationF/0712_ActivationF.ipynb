```python
import numpy as np
import matplotlib.pyplot as plt

#step function
def step_function(x):
    y= x>0
    return (y)

x= np.arange(-5.0, 5.0, 0.1)
y= step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

x= np.arange(-5.0, 5.0, 0.1)
y= sigmoid(x)
plt.plot(x,y)
plt.plot([0,0], [1.0, 0.0], ':') # 가운데 점선 추가
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
x= np.arange(-5.0, 5.0, 0.1)
y=np.tanh(x)

plt.plot(x,y)
plt.plot([0,0], [1.0, -1.0], ':') # 가운데 점선 추가
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#ReLu function
def relu(x):
    return np.maximum(0, x)

x= np.arange(-5.0, 5.0, 0.1)
y= relu(x)

plt.plot(x,y)
plt.plot([0,0], [5.0, 0.0], ':') # 가운데 점선 추가
plt.title('Relu Function')
plt.show()
```

```python
x= np.arange(-5.0, 5.0, 0.1)
y= np.exp(x)/np.sum(np.exp(x))

plt.plot(x,y)
plt.title('Softmax Function')
plt.show()
```
